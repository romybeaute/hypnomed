# Map the informations of our different data sets to correlate Phenomenological data with fc-IRMf data

*And plot the different structures of the datasets for clarity purposes*

```python
import pandas as pd
import numpy as np 
import os

data_path = '/home/romy.beaute/projects/hypnomed/data/participants'
```

```python
df_composite_score = pd.read_pickle(os.path.join(data_path,'composite_score.pkl'))
df_composite_score.head()
```

```python
df_hypnomed = pd.read_csv(os.path.join(data_path,'hypnomed.csv'),sep=';')
subs_order = df_hypnomed['sub_order'].tolist() #order of subjects wrt to their scan session
df_hypnomed.head()
```

```python
df_phenoscore = pd.read_csv('/home/romy.beaute/projects/hypnomed/analysis/pheno_stats/questionnaires_hypnomed.csv')

df_phenoscore.head()
```

```python
df_pheno = pd.read_csv('/home/romy.beaute/projects/hypnomed/analysis/pheno_stats/Datas_Prisca.csv',sep=';')
df_pheno = df_pheno.loc[df_pheno['Sujet'].isin(subs_order)]
dict_id_order = dict(zip(df_hypnomed.sort_values('sub_order').sub_order,df_hypnomed.sort_values('sub_order').sub_id))
reordered_subs = df_hypnomed.sort_values('sub_order').sub_id.tolist() #sorting the sub-0x wrt to the sub_order to retrieve mapping between df_pheno and irmf data
reordered_subs = list(np.repeat(reordered_subs,2)) #x2 because we get the value for each block condition (meditation and hypnosis only) for each subject 
df_pheno['Sub-id']=reordered_subs
df_pheno['group']=list(np.repeat(df_phenoscore['group'],2))
df_pheno = df_pheno.sort_values('Sub-id')
df_pheno['id']=list(np.repeat(df_phenoscore['id'],2))
keep_cols = ['Sujet', 'Sub-id','group','Bloc', 'states','difficulty_delta', 'Pheno_Etat_delta','id','age','genre','Hypnos_exp', 'Medit_exp'] #keep only the informations of interest, for clarity purpose
df_pheno = df_pheno[keep_cols]
df_pheno.to_csv(os.path.join(data_path,'data_pheno.csv'))
df_pheno
```

```python
print(' - G1 : meditation in block 2, hypnose in block 3\n - G2 : meditation in block 3, hypnose in block 2')
df_pheno_meditation = df_pheno[df_pheno['states']=='Meta_awareness']
print(' - n={} subjects in meditation'.format(len(df_pheno_meditation)))

medit_Difficulty_dict = dict(zip(df_pheno_meditation['Sub-id'],df_pheno_meditation['difficulty_delta']))
medit_PhenoEtat_dict = dict(zip(df_pheno_meditation['Sub-id'],df_pheno_meditation['Pheno_Etat_delta']))

df_pheno_meditation.head()
```

```python
print(' - G1 : meditation in block 2, hypnose in block 3\n - G2 : meditation in block 3, hypnose in block 2')
df_pheno_hypnose = df_pheno[df_pheno['states']=='Absorption']
print(' - n={} subjects in hypnose'.format(len(df_pheno_hypnose)))

hypno_Difficulty_dict = dict(zip(df_pheno_hypnose['Sub-id'],df_pheno_hypnose['difficulty_delta']))
hypno_PhenoEtat_dict = dict(zip(df_pheno_hypnose['Sub-id'],df_pheno_hypnose['Pheno_Etat_delta']))

df_pheno_hypnose.head()
```
